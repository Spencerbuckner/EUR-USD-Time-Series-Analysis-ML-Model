### 4/14/2017 was deleted from set after being saved. Market was closed due to Good Friday so all values were set to 0.

### Values set to 0 in the 1 mo and 3 mo columns will be set to 0.01 as well. Since the dataset only goes to 2 decimals, andthing below was set to absolute 0, and will need to be updated since they are not actually 0.

```python
def zero_fix(num):
    if num == 0:
        return 0.01
    else:
        return num
```

```python
df_concat['1 mo'] = df_concat['1 mo'].map(zero_fix)
df_concat['3 mo'] = df_concat['3 mo'].map(zero_fix)
```
