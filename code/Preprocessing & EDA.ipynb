```python
df_merged = fixed_days_df.merge(yield_df, how='left', left_index=True, right_index=True)
df_merged
```

```python
df_merged.columns = [i.replace(" ", "_") for i in df_merged.columns]
```

```python
df_merged.columns
```

```python
df_merged.head()
```

```python
df_merged.to_csv('../data/datamerged.csv')
```

```python
model_df = pd.DataFrame(index=df_merged.index)

for i in range(2,31):
    for col in df_merged.columns:
        model_df[f'{col}_rolling_avg_{i}'] = df_merged[col].rolling(window=i, min_periods=0).mean().shift(1)
        model_df[f'{col}_rolling_avg_std_{i}'] = df_merged[col].rolling(window=i, min_periods=0).std().shift(1)

#target
model_df['Target'] = df_merged['BO']

# get rid of the trend
model_df = model_df.diff() # or diff

model_df.dropna(inplace=True)
model_df
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
model_df.columns[:10]
```

```python
for i in model_df.columns[:10]: # only looking at first 10
    plt.title(f'{i} Autocorrelation')
    plot_acf(model_df[i], lags=30);
```

```python
for i in model_df.columns[:10]: # only looking at first 10
    plt.title(f'{i} Partial Autocorrelation')
    plot_pacf(model_df[i], lags=30);
```
