```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/modeling_data.csv')
df.drop(columns='Unnamed: 0', inplace=True)
df.head()
```

```python
df['Date_Time'] = pd.to_datetime(df['Date_Time'])
df.set_index('Date_Time', inplace=True)
df.sort_index(inplace=True)
df.head()
```

```python
df['tommorow_BC'] = df['BC'].shift(-1)
df.dropna(inplace=True)
df.head(2)
```

```python
df.corr()[['tommorow_BC']].sort_values('tommorow_BC', ascending=False)[1:]
```

```python
model_df_corr = df.corr()[['tommorow_BC']].sort_values('tommorow_BC', ascending=False)[1:]
plt.figure(figsize=(10, 10))
sns.heatmap(model_df_corr, cmap='coolwarm', vmin=-1, vmax=1, annot=True, annot_kws={"size":15})
plt.title('Correlations to Tommorow Bid Close Price', size=12);
plt.savefig('../visualizations/tommorow_bid_close_corr_heatmap.png')
```
