```python
import pandas as pd
import numpy as np
import requests
import time

from bs4 import BeautifulSoup
```

```python
month = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
day1 = list(range(1,32))
day = [str(x) for x in day1]
year1 = list(range(2007, 2021))
year = [str(x) for x in year1]
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from requests.packages.urllib3.util.retry import Retry
```

```python
new_list = []
for y in year:
    for m in month:
        for d in day:
            new_list.append(f'https://www.forexfactory.com/calendar?day={m}{d}.{y}')
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
df = []
for i in new_list:
    try:
        driver.get(i) # f string
        # Get the table
        table = driver.find_element(By.CLASS_NAME, "calendar__table")
        # Iterate over each table row
        for row in table.find_elements(By.TAG_NAME, "tr"):
            # list comprehension to get each cell's data and filter out empty cells
            row_data = list(filter(None, [td.text for td in row.find_elements(By.TAG_NAME, "td")]))
            if row_data == []:
                continue
            row_data.append(i[-4:] + i[-10:-5])
            df.append(row_data)
            print(i[-4:] + i[-10:-5])
                       
    except Exception as e:
        print(e)
    time.sleep(60)
driver.quit()
```

```python
pd.DataFrame(df)
```
